## Task 1 — FizzBuzz (a classic!)

Loop through numbers 1 to 30 and for each number print:
- `"Fizz"` if it is divisible by 3
- `"Buzz"` if it is divisible by 5
- `"FizzBuzz"` if it is divisible by **both** 3 and 5
- the number itself otherwise

*Hint: check the "both" case first!*

````python
total = 0

for n in range(1, 30):
    if n % 3 == 0:
        print(f"{n} = fizz")
    if n % 5 == 0:
        print(f"{n} = Buzz")
    if n % 3 == 0 and n % 5 == 0:
        print(f"{n} = fizzBuzz")

````

## Task 2 — Sum of even numbers

Use a loop to calculate the sum of all **even** numbers between 1 and 50 (inclusive). Print the result.

````python
total = 0

for n in range(1, 51):
    if n % 2 == 0:
        total = total + n

print("Sum of even numbers:", total)
````

## Task 3 — Menu-driven input validation

Print this menu:
```
1. Say Hello
2. Say Goodbye
3. Exit
```
Ask the user to choose (`input()`). Using a `while` loop:
- if they enter `1` → print `"Hello!"`
- if they enter `2` → print `"Goodbye!"`
- if they enter `3` → print `"Exiting..."` and `break` out of the loop
- anything else → print `"Invalid choice, try again"`

The menu should keep repeating until the user chooses 3.

*This is a preview of tomorrow's project style — a program that keeps running until the user decides to stop!*

````python
print ("Say Hello")
print ("Say Goodbye")
print("exit")

while True:
    number = int(input("Enter a number: "))
    if number == 1:
       print("Say Hello")
    elif number == 2:
       print ("Say Goodbye")
    elif number == 3:
       print ("Exiting")
       break
    else:
        print ("invalid choice, try again")
````
